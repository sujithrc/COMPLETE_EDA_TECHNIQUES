```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
#merging both train and test set 
#using append or merge(leftjoin, rightjoin) or concate

#df= df_train.append(df_test)
#df.head()


# Assuming df_train and df_test are DataFrame objects
df = pd.concat([df_train, df_test], ignore_index=True)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.keys()
```

```python
df.drop(['User_ID'], axis=1, inplace=True)
df.head()
```

```python
# coverting gender column into categorical to numerical

df['Gender']= df['Gender'].map({'F':0,"M":1})
df.head()
```

```python
#handling categorial feature age

df['Age'].unique()
```

```python

# Get the value counts of the 'Age' column
age_counts = df['Age'].value_counts()

# Display the details about the 'Age' column
print(age_counts)


```

```python
#pd.get_dummies(df['Age'],drop_first=True)

df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head()
```

```python
# Define the mapping of age categories to numerical values
age_mapping = {'0-17': 1, '18-25': 2, '26-35': 3, '36-45': 4, '46-50': 5, '51-55': 6, '55+': 7}

# Apply the mapping to the 'Age' column
df['Age'] = df['Age'].map(age_mapping)

# Display the DataFrame after mapping
df.head()
```

```python
city = df['City_Category'].value_counts()
city
```

```python
# method 1: use "label encoder" to convert categorical value into numerical value
# method: mapping


# Define the mapping of categorical values to numerical values
city_mapping = {'A': 1, 'B': 2, 'C': 3}

# Apply the mapping to the 'City_Category' column
df['City_Category'] = df['City_Category'].map(city_mapping)

# Display the updated DataFrame
print(df.head())

```
