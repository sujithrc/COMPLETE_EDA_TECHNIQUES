```python
import pandas as pd

# Read the datasets
train_df = pd.read_excel('Data_Train.xlsx')
test_df = pd.read_excel('Test_set.xlsx')

# Add a 'Price' column to test_df filled with NaNs to match train_df columns
#test_df['Price'] = float('nan')

# Concatenate train_df and test_df
final_df = pd.concat([train_df, test_df], ignore_index=True)

# Display the combined DataFrame
final_df.head()
```

```python
# we are gonna convert the duration object value into int values

import pandas as pd

# Sample DataFrame
# final_df = your_dataframe_here

# Define function to convert duration to minutes
def duration_to_minutes(duration):
    duration_split = duration.split()
    total_minutes = 0
    for part in duration_split:
        if 'h' in part:
            total_minutes += int(part[:-1]) * 60
        elif 'm' in part:
            total_minutes += int(part[:-1])
    return total_minutes

# Apply the function to the Duration column
final_df['Duration_minutes'] = final_df['Duration'].apply(duration_to_minutes)

# Display the DataFrame with the new Duration_minutes column
print(final_df.head())
```
